```python
def ysymarr(arr):
    C = np.vstack((np.flipud(arr),arr))
    return C
```

```python
def fig1(ns=None):
    # To draw number/mass density from simulations with different cooling  prescribtion.
    # And SiO emission maps for a single line with i = 0 for these cooling models. [2 figures] 
    if ns == None : ns = 55
    wds = ['HighRes_adiab1010/','HighRes_powlawT10/','HighRes_atm1010/','HighRes_Tabc1010/','HighRes_molc1010/']
    titL = [r'Nil (Adiabatic)',r'Power law', r'Atomic',r'Tabulated',r'Molecular']
    DtL = []
    for i in wds:
        DtL.append(pp.pload(ns,w_dir=i))


    f1 = plt.figure(figsize=[20,6])
    plt.subplots_adjust(left=0.05,right=0.95,wspace=0.2)
    for i in range(len(wds)):
        ax = f1.add_subplot(1,5,i+1)
        ext = [-DtL[i].x1.max(),DtL[i].x1.max(),DtL[i].x2.min(),DtL[i].x2.max()]
        svar = ysymarr(np.log10(DtL[i].rho))
        im = ax.imshow(svar.T,extent=ext,origin='image')
        cb = plt.colorbar(im,orientation='vertical')
        ax.set_xlabel(r'R')
        ax.set_ylabel(r'Z')
        ax.set_title(titL[i])
        plt.minorticks_on()
```

```python
def fig2(wds=None):
    # To draw 5 arrays for a single time step -- 
    # H fractions of Molecular cooling with T and Mu, 
    # SiO Emission maps for Molecular cooling all lines (1->0,2->1,3->2,5->4,8->7) (i=30), 
    # SiO Emission maps for Tabulated cooling All lines (1->0,2->1,3->2,5->4,8->7) (i=0) [3 figures ]
    
    if wds == None : wds = os.getcwd()+'/'
    D = pp.pload(55,w_dir=wds)
    varL = [np.log10(D.rho*D.fHI), np.log10(D.rho*D.fH2), np.log10(D.rho*D.fHII), np.log10(D.T), D.MU]
    titL = [r'log($\rho_{HI}$)',r'log($\rho_{H2}$)', r'log($\rho_{HII}$)',r'log(T)',r'$\mu$']
    vminL = [np.log10(50.0) if i == 3 else varL[i].min() for i in range(len(varL))]
    vminL[2] = np.log10(1.0e-4)

    f1 = plt.figure(figsize=[20,6])
    plt.subplots_adjust(left=0.05,right=0.95,wspace=0.2)
    for i in range(len(varL)):
        ax = f1.add_subplot(1,5,i+1)
        ext = [-D.x1.max(),D.x1.max(),D.x2.min(),D.x2.max()]
        svar = ysymarr(varL[i])
        im = ax.imshow(svar.T,extent=ext,origin='image',vmin=vminL[i])
        cb = plt.colorbar(im,orientation='vertical')
        ax.set_xlabel(r'R')
        ax.set_ylabel(r'Z')
        ax.set_title(titL[i])
        plt.minorticks_on()
```

```python
def fig3(arr1=None, arr2=None, line1=[None,None]):
    # To draw Intensity maps, PV diagram and Spectrum. -- 
    #For one single line with i = 0, 30, 45 -- Total 3 figures. 
    
    
    fig = plt.figure(figsize=[12,15])
    # SiO Intensity Map.
    ax1 = plt.subplot2grid((2,2), (0,0))
    if arr1 == None : arr1 = random.randint(5,size=(15,30))
    im1 = ax1.imshow(arr1.T,origin='image')
    cb1 = plt.colorbar(im1,orientation='vertical', use_gridspec=True)
    ax1.set_xlabel(r'Radius [AU]')
    ax1.set_ylabel(r'Height [AU]')
    ax1.set_aspect('equal')
    plt.plot(linspace(10,10,100),linspace(0.0,29.0,100),'w') # cut to represent where PV diagram is drawn.
    plt.plot(linspace(0,14,100),linspace(24.0,24.0,100),'k') # Cut to represent where spectrum is taken.
    ax1.axis([0.0,14.0,0.0,29.0])

    # Position Velocity plot
    ax2 = plt.subplot2grid((2,2), (0,1))
    if arr2 == None: arr2 = random.randint(5,size=(15,30))
    im2 = ax2.imshow(arr2.T,origin='image')
    cb2 = plt.colorbar(im2,orientation='vertical', use_gridspec=True)
    ax2.set_xlabel(r'Velocity [km s$^{-1}$]')
    ax2.set_ylabel(r'Position [AU]')
    ax1.set_aspect('equal')

    # Spectrum.
    ax3 = plt.subplot2grid((2,1), (1,0), rowspan=2)
    if line1[0] == None : line1[0] = linspace(-20.0,20.0,100)
    if line1[1] == None : line1[1] = exp(-(line1[0]**2)/(2.0*pi*0.7**2))
    plt.plot(line1[0], line1[1] , linestyle='steps-mid')
    ax3.minorticks_on()
    ax3.set_xlabel(r'Velocity [km s$^{-1}$]')
    ax3.set_ylabel(r'Intensity [K]')

    fig.tight_layout()
```

```python
def fig4():
    # To draw plot of fractional abundance with velocity for SiO used for estimating the SiO Intensity. 
    f1 = plt.figure(figsize=[10,10])
    ax1 = f1.add_subplot(111)
    ax1.set_xlabel(r'Velocity [km s$^{-1}$]')
    ax1.set_ylabel(r'Fractional Abundance [$X_{\rm SiO}$]')
    v = linspace(0.0,125.0,500)
    Xdum = -2.48e-8*v**5 + 5.50e-6*v**4 - 4.28e-4*v**3 + 1.24e-2*v**2 + 2.52e-2*v - 1.20e1
    X = 10**(Xdum)
    ax1.semilogy(v,X)
    ax1.axis([0.0,125.0,1e-25,1.0e-3])
    ax1.minorticks_on()
    
```
